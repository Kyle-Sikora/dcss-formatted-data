```python
import pandas as pd
from functools import reduce
import matplotlib.pyplot as plt
import zipfile
import numpy as np

print("Loading Data")
df = pd.read_csv('dcss_playtime.csv')
print("Data Loaded")
```

```python
plt.xscale("log")
binss=np.logspace(0.1, 10, 50, endpoint=True)
print(binss)
plt.hist(df['playtime'], bins=binss)
plt.title("Histogram of player total playtime")
plt.xlabel('seconds')
plt.ylabel('count of players')
plt.show()
```

```python
df = df.sort_values(by=['playtime'], ascending=True)
print(str(len(df['playtime'])) + " players")

y_axis = []

for i in range(0,len(df['playtime'])):
    y_axis.append(i*1.0/len(df['playtime']))

plt.plot(df['playtime'],y_axis,label="median")
plt.xlabel('time played (seconds)')
plt.xscale("log")
plt.ylabel('player percentile')
plt.title("Player Total Playtime Percentile")
plt.show()
```
